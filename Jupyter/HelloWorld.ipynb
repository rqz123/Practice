```python
import os

data_dir = os.getcwd() + '\\Data\\Train'
print(data_dir)
```

```python
import pathlib
from PIL import Image

data_dir = pathlib.Path(data_dir)
print(data_dir)

image_count = len(list(data_dir.glob('*/*.bmp')))
print('image_count: ', image_count)

for images in data_dir.glob('*.bmp'):
    try:
        im = Image.open(images)
        im.verify()
    except Exception:
        print(images)
        print('Exception')
    except SyntaxError:
        print(images)
        print('SyntaxError')
    except UnidentifiedImageError:
        print(images)
        print('UnidentifiedImageError')

print('All images have been verified!')
```

```python
"""
Simple demo of a scatter plot
"""
import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
